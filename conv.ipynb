```python
# 單筆資料
from pandas.io.json import json_normalize
with open('data/7-11/台中市.json',encoding='utf-8') as json_file:
    x = json.load(json_file)
    result = json_normalize(x, 'stores', ['city_id', 'city_name'])
    # result.to_csv('data/con_json.csv')
    print(result.head(3))
```

```python
# 讀資料夾內資料
path = 'data/7-11'

json_files = [i for i in os.listdir(path)]
print(json_files)
# json_files = []
# for i in os.listdir(path):
#     json_files.append(i)
#     print(i)
```

```python
for i in json_files:
    with open('data/7-11/'+i,encoding='utf-8') as json_file:
        x = json.load(json_file)
        result = json_normalize(x, 'stores', ['city_id', 'city_name'])
        result.to_csv('data/total_con_json.csv',mode='a',encoding='cp950')
        print(result.head(3))
```
